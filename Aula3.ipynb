```python
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv('diabetes.csv')
```

```python
df.value_counts('Outcome')
```

```python
sns.kdeplot(data=df,x='Age')
```

```python
df['Resultado'] = ''
df.loc[df['Outcome']== 1,'Resultado'] = 'Diabetico'
df.loc[df['Outcome']== 0,'Resultado'] = 'Não Diabetico'
sns.countplot(data = df, x = 'Resultado', palette='RdYlBu')
```

```python
sns.scatterplot(data = df,y='Insulin',x='Glucose')
```
